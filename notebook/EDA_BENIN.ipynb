```python
#check for outliers, especially in sensor readings (ModA, ModB) and wind speed data (WS, WSgust).
# Calculate Z-scores for the column
z_scores = stats.zscore(data['ModA'])

# Define threshold for outliers (e.g., |z| > 3)
threshold = 3
outliers_z = data[abs(z_scores) > threshold]
outliers_z
```

```python
# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data.index, data['ModA'])
plt.xlabel('Index')
plt.ylabel('ModA')
plt.show()
```
